```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications import vgg19
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
img_nrows = 400
img_ncols = 400
```

```python
def preprocess_image(image_path):
    img = load_img(image_path, target_size=(img_nrows, img_ncols))
    img = img_to_array(img)
    img = np.expand_dims(img, axis=0)
    return vgg19.preprocess_input(img)

def deprocess_image(x):
    x = x.reshape((img_nrows, img_ncols, 3))
    x[:, :, 0] += 103.939
    x[:, :, 1] += 116.779
    x[:, :, 2] += 123.68
    x = x[:, :, ::-1]  # BGR -> RGB
    return np.clip(x, 0, 255).astype('uint8')
```

```python
content_layer = 'block5_conv2'
style_layers = [
    'block1_conv1', 'block2_conv1',
    'block3_conv1', 'block4_conv1',
    'block5_conv1'
]


content_weight = 1e3
style_weight = 1e-2
```

```python
content_path = 'content.jpg'
style_path = 'style.jpg'

content_image = preprocess_image(content_path)
style_image = preprocess_image(style_path)
```

```python
def get_model():
    vgg = vgg19.VGG19(include_top=False, weights='imagenet')
    vgg.trainable = False
    outputs = [vgg.get_layer(name).output for name in style_layers + [content_layer]]
    return Model([vgg.input], outputs)

model = get_model()
```

```python
def get_feature_representations(model):
    style_outputs = model(style_image)
    content_outputs = model(content_image)
    style_features = [style_layer for style_layer in style_outputs[:len(style_layers)]]
    content_feature = content_outputs[-1]
    return style_features, content_feature

def gram_matrix(x):
    x = tf.squeeze(x)
    x = tf.reshape(x, (-1, x.shape[-1]))
    return tf.matmul(x, x, transpose_a=True)

def compute_loss(model, loss_weights, init_image, gram_style_features, content_feature):
    input_tensor = tf.concat([init_image], axis=0)
    features = model(input_tensor)
    
    style_output_features = features[:len(style_layers)]
    content_output_feature = features[-1]

    style_score = 0
    content_score = 0

    for target, comb in zip(gram_style_features, style_output_features):
        gram_comb = gram_matrix(comb)
        style_score += tf.reduce_mean(tf.square(gram_comb - target))

    content_score = tf.reduce_mean(tf.square(content_output_feature - content_feature))
    
    style_score *= loss_weights[1]
    content_score *= loss_weights[0]

    loss = style_score + content_score
    return loss

@tf.function()
def compute_grads(cfg):
    with tf.GradientTape() as tape:
        loss = compute_loss(**cfg)
    return tape.gradient(loss, cfg['init_image']), loss
```

```python
init_image = tf.Variable(content_image, dtype=tf.float32)

opt = tf.optimizers.Adam(learning_rate=5.0)

style_features, content_feature = get_feature_representations(model)
gram_style_features = [gram_matrix(style_feature) for style_feature in style_features]

loss_weights = (content_weight, style_weight)
cfg = {
    'model': model,
    'loss_weights': loss_weights,
    'init_image': init_image,
    'gram_style_features': gram_style_features,
    'content_feature': content_feature
}
```

```python
epochs = 10
steps_per_epoch = 100

for n in range(epochs):
    for i in range(steps_per_epoch):
        grads, loss = compute_grads(cfg)
        opt.apply_gradients([(grads, init_image)])
        init_image.assign(tf.clip_by_value(init_image, -127.5, 127.5))
    print(f"Epoch {n + 1}, loss={loss}")
```

```python
final_img = deprocess_image(init_image.numpy())
plt.imshow(final_img)
plt.axis('off')
plt.show()
```
